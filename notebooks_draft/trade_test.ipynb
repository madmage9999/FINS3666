```python
import pandas as pd
import numpy as np
```

```python
month = '2023_12'
result_dir = f'trade_result_2/{month}'
```

```python
df = pd.read_csv(f'{result_dir}/{month}_tradebook.csv')

def avg_price(row):
    return pd.Series([sum(row.back_trades['p'])/len(row.back_trades['p']) if len(row.back_trades['p']) > 0 else 0, sum(row.lay_trades['p'])/len(row.lay_trades['p']) if len(row.lay_trades['p']) > 0 else 0])

def total_volume(row):
    return pd.Series([sum(row.back_trades['v']), sum(row.lay_trades['v'])])
```

```python
import ast
df['back_trades'] = [ast.literal_eval(x) for x in df['back_trades']]
df['lay_trades'] = [ast.literal_eval(x) for x in df['lay_trades']]

df[['avg_back_price', 'avg_lay_price']] = df.apply(lambda row: avg_price(row), axis=1, result_type='expand')
df[['back_vol', 'lay_vol']] = df.apply(lambda row: total_volume(row), axis=1, result_type='expand')
df
```

```python
dec = '2023_12'
jan = '2024_01'
feb = '2024_02'

df1 = pd.read_csv(f'trade_result_2/{dec}/{dec}_summary.csv',
                 dtype={'market_id': 'string', 'selection_id': 'string'},
                 )
df2 = pd.read_csv(f'trade_result_2/{jan}/{jan}_summary.csv',
                 dtype={'market_id': 'string', 'selection_id': 'string'},
                 )
df3 = pd.read_csv(f'trade_result_2/{feb}/{feb}_summary.csv',
                 dtype={'market_id': 'string', 'selection_id': 'string'},
                 )
summary = pd.concat([df1, df2, df3], axis=0)
summary
```

```python
summary['gpl'] = summary['back_return'] + summary['lay_return']
summary['stake'] = summary['back_v_sum'] + summary['lay_v_sum']
```

```python
import plotly.express as px

def bet_eval_metrics(d):

    metrics = pd.DataFrame(d
    .agg({"gpl": "sum", "stake": "sum"})
    ).transpose().assign(pot=lambda x: x['gpl'] / x['stake'])

    return(metrics[metrics['stake'] != 0])

# Cumulative PL by market to visually see trend and consistency
def bet_eval_chart_cPl(d):

    d = (
        d
        .groupby('market_id')
        .agg({'gpl': 'sum'})
    )

    d['market_number'] = np.arange(len(d))
    d['cGpl'] = d.gpl.cumsum()

    chart = px.line(d, x="market_number", y="cGpl", title='Cumulative Gross Profit', template='simple_white')

    return(chart)
```

```python
eval_df = bet_eval_metrics(summary)
print(f'Total Gross Profit/Loss % = {(eval_df["gpl"]/eval_df["stake"]*100).values[0]}')
eval_df
```

```python
bet_eval_chart_cPl(summary)
```
