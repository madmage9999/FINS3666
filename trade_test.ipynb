```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test_tradebook.csv')

def avg_price(row):
    return pd.Series([sum(row.back_trades['p'])/len(row.back_trades['p']), sum(row.lay_trades['p'])/len(row.lay_trades['p'])])

def total_volume(row):
    return pd.Series([sum(row.back_trades['v']), sum(row.lay_trades['v'])])
```

```python
import ast
df['back_trades'] = [ast.literal_eval(x) for x in df['back_trades']]
df['lay_trades'] = [ast.literal_eval(x) for x in df['lay_trades']]

df[['avg_back_price', 'avg_lay_price']] = df.apply(lambda row: avg_price(row), axis=1, result_type='expand')
df[['back_vol', 'lay_vol']] = df.apply(lambda row: total_volume(row), axis=1, result_type='expand')
df
```
